### –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞—Å—á–µ—Ç–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è

–§–æ—Ä–º—É–ª–∞ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞

$$
Recovery = {C x (F-T)\over F x (C-T)}
$$

–í –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ —Ä–æ–ª—å –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –±—É–¥—É—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ —Å—Ç–æ–ª–±—Ü—ã:<br><br>
**Recovery** - *rougher.output.recovery*<br>
**C** - *rougher.output.concentrate_au*<br>
**F** - *rougher.input.feed_au*<br>
**T** - *rougher.output.tail_au*<br>

```python
df_train['calc_recovery'] =\
(df_train['rougher.output.concentrate_au'] * (df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au']))/\
(df_train['rougher.input.feed_au'] * (df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au']))*100
```

```python
#df_train.head()
```

```python
mae_recovery = mean_absolute_error(df_train['rougher.output.recovery'],df_train['calc_recovery']) 
print(f'MAE = {mae_recovery:.5f}')
```

–ò–∑ –≥—Ä–∞—Ñ–∏–∫–∞ –≤–∏–¥–Ω–æ, —á—Ç–æ –µ—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–∏—Å–∏–º—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤

–†–∞—Å—á–µ—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∏–Ω—Ñ–ª—è—Ü–∏–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
def calc_vif(df_f,drop_param):
    """
    –†–∞—Å—á–µ—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    """
    #–£–¥–∞–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å –≤—ã—Å–æ–∫–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–µ–π
    df_f = df_f.drop([drop_param],axis=1)
    pp = df_f
    X = add_constant(pp)  
    #–ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –Ω–æ–≤–æ–≥–æ Series —Å –Ω–æ–≤—ã–º–∏ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è–º–∏
    ds=pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])], index=X.columns)  
    return ds,df_f
```

–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ Series –≤ Dataframe

```python
def conv_ser_to_df(ser):
    t = ser.to_frame().reset_index()
    t = t.rename(columns= {0: 'coef'})
    #–£–¥–∞–ª–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏ 0 c –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Ä–∞—Å—á–µ—Ç–æ–≤
    t = t.drop(index=0)
    return t
```

```python
df_f = df_test.drop(['date'],axis=1)
```

–†–∞—Å—á–µ—Ç –Ω–∞—á–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è –∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
pp = df_f
X = add_constant(pp)  
#–ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –Ω–æ–≤–æ–≥–æ Series —Å –Ω–æ–≤—ã–º–∏ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è–º–∏
ds=pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])], index=X.columns)
```

```python
tind = conv_ser_to_df(ds)
```

```python
tind = tind.sort_values(by='coef',ascending=False)
```

## –ú–æ–¥–µ–ª—å

**–§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE**

```python
def smape_func(target, pred):
    '''
    –ú–µ—Ç–æ–¥ –ø–æ–ª—É—á–µ–Ω–∏—è sMape –¥–ª—è –æ–¥–Ω–æ–≥–æ —ç—Ç–∞–ø–∞
    '''
    
    smape = abs(target-pred)/((abs(target)+abs(pred))/2)*100
    #smape = smape.fillna(value=0)

    smape = np.mean(smape)
    return smape
```

```python
def summary_smape(smape_rouger, smape_final):
    
    '''
    –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏—Ç–æ–≥–æ–≤–æ–π sMAPE –¥–ª—è smape_rouger –∏ smape_final
    '''
    
    # –∏—Ç–æ–≥–æ–≤—ã–π smape
    summary_smape = 0.25 * smape_rouger + 0.75 * smape_final
    
    return summary_smape
```

#### –ü—Ä–æ–≤–µ—Ä–∫–∞ –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ

```python
def best_model(features_train, target_train, features_test, depth, estimators):
    # –º–µ—Ç–æ–¥ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ —Å–ª—É—á–∞–π–Ω–æ–º –ª–µ—Å–µ

    # –∏–Ω–∏—Ü–∏–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å —Å –ª—É—á—à–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
    model = RandomForestRegressor(random_state=12345, max_depth = depth, n_estimators=estimators)
    # –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å
    model.fit(features_train, target_train)
    # –ø–æ–¥–≥–æ—Ç–æ–≤–∏–º –ø—Ä–æ–≥–Ω–æ–∑ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
    pred_test_rf = model.predict(features_test)

    return pred_test_rf
```

–†–∞—Å—á–µ—Ç —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –Ω–∞ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–π –º–æ–¥–µ–ª–∏ 

```python
def dumm_model(features_train, target_train, features_test):
    """
    –º–µ—Ç–æ–¥ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–π –º–æ–¥–µ–ª–∏ 
    """
    # –∏–Ω–∏—Ü–∏–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å —Å –ª—É—á—à–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
    model_dummy = DummyRegressor(strategy="mean")
    model_dummy.fit(features_train, target_train)
    #–ø—Ä–æ–≥–Ω–æ–∑ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
    pred_test = model_dummy.predict(features_test)
    return pred_test
```
